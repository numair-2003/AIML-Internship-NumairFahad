### 2.3 Min-Max Normalization

A common preprocessing step in ML is scaling values into a fixed range, typically [0, 1]. The formula for min-max normalization is:

$$ x_{norm} = \frac{x - min(x)}{max(x) - min(x)} $$

**Exercise:** Implement `normalize(x)`, which takes a NumPy array and returns it min-max normalized. Use `np.min()` and `np.max()` — no loops.

````python
def normalize(x):
    ### YOUR CODE HERE ###
    normalized_x = (x - np.min(x)) / (np.max(x) - np.min(x))
    return normalized_x
    ### END ###
    pass

test_scores = np.array([40, 55, 70, 85, 100])
print(normalize(test_scores))
````

**Expected Output:**
```
[0.   0.25 0.5  0.75 1.  ]
```

### 2.4 Reshaping Arrays

`array.shape` tells you an array's dimensions, and `array.reshape(...)` lets you rearrange it into a different shape (as long as the total number of elements matches).

**Exercise:** Given the flat array `readings` (12 sensor readings), reshape it into a 3x4 matrix called `readings_matrix`, then compute the mean of each row using `axis=1`.

### 2.5 Broadcasting

Broadcasting lets NumPy apply an operation between arrays of different shapes without manually looping — for example, subtracting a single row of column-averages from every row of a matrix.

**Exercise:** Using `readings_matrix` from the previous exercise, subtract the column-wise mean (`axis=0`) from every row, producing `centered_matrix`.

````python
### YOUR CODE HERE ###
centered_matrix = readings_matrix - readings_matrix.mean(axis=0)
### END ###

print(centered_matrix)
````

### 2.6 Why Vectorization Matters

Run the cell below as-is to compare a manual loop against NumPy's vectorized `np.dot()` for computing a dot product on a larger array. Note the difference in computation time.

````python
import time

a = list(range(1_000_000))
b = list(range(1_000_000))

# Manual loop
tic = time.process_time()
loop_dot = 0
for i in range(len(a)):
    loop_dot += a[i] * b[i]
toc = time.process_time()
print("Loop result:", loop_dot, "| Time:", (toc - tic) * 1000, "ms")

# Vectorized with NumPy
a_np = np.array(a)
b_np = np.array(b)

tic = time.process_time()
vec_dot = np.dot(a_np, b_np)
toc = time.process_time()
print("Vectorized result:", vec_dot, "| Time:", (toc - tic) * 1000, "ms")
````

**What to remember from Part 2:**
- NumPy arrays support vectorized operations — no loops needed for element-wise math
- `.reshape()` changes an array's dimensions without changing its data
- Broadcasting lets you operate between arrays of different (but compatible) shapes
- Vectorized NumPy code is dramatically faster than an equivalent manual loop

## Part 3 — Pandas Essentials

Pandas' main object is the **DataFrame** — a table of rows and columns, similar to a spreadsheet, but fully controllable through code.

### 3.1 Creating and Exploring a DataFrame

````python
import pandas as pd

data = {
    "product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Webcam"],
    "price": [1200, 25, 45, 300, 60],
    "units_sold": [15, 120, 95, 40, 70]
}
df = pd.DataFrame(data)

df.head()
````

### 3.2 Filtering & Creating New Columns

**Exercise:** Create a new column `revenue` equal to `price * units_sold`. Then create a filtered DataFrame `high_revenue` containing only rows where `revenue` is above 5000.

````python
### YOUR CODE HERE ###
df["revenue"] = df["price"] * df["units_sold"]
high_revenue = df[df["revenue"] > 5000]
### END ###

print(high_revenue)
````

### 3.3 Grouping & Aggregating

**Exercise:** Using the `products_extended` DataFrame below (which has a `category` column), compute the average `price` and total `units_sold` per category, in a single `.groupby().agg()` call.

````python
products_extended = pd.DataFrame({
    "product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Webcam", "Desk", "Chair"],
    "category": ["Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Furniture", "Furniture"],
    "price": [1200, 25, 45, 300, 60, 250, 180],
    "units_sold": [15, 120, 95, 40, 70, 20, 35]
})

### YOUR CODE HERE ###
category_summary = products_extended.groupby("category").agg(
    average_price = ("price", "mean"),
    total_units_sold = ("units_sold", "sum")
)
### END ###

print(category_summary)
````

### 3.4 Handling Missing Values

**Exercise:** The DataFrame below has missing values in the `price` column. Create `df_filled`, where missing prices are replaced with the column's average price, using `.fillna()`.

````python
import numpy as np

messy_data = pd.DataFrame({
    "product": ["Tablet", "Charger", "Case", "Earbuds"],
    "price": [400, np.nan, 20, np.nan]
})

### YOUR CODE HERE ###
df_filled = messy_data.fillna(messy_data["price"].mean())
### END ###

print(df_filled)
````

**What to remember from Part 3:**
- A DataFrame is a table you can filter, transform, and summarize entirely through code
- New columns can be created directly from existing ones
- `.groupby().agg()` is the standard way to summarize data by category
- `.fillna()` is one common way to handle missing values — the right strategy depends on the situation
